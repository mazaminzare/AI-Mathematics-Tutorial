## Time Series Analysis

Time series analysis involves the study of data points collected or recorded at specific time intervals. This tutorial covers key concepts, mathematical background, and numerical examples.

### 1. Components of a Time Series

Time series data typically consist of several components:
1. **Trend**: The long-term progression of the series.
2. **Seasonality**: Regular patterns or cycles of variation.
3. **Cyclic**: Irregular, longer-term fluctuations.
4. **Irregular (Noise)**: Random variations.

*Example:*

Suppose we have monthly sales data for a company over three years:

| Month | Sales |
|-------|-------|
| Jan   | 100   |
| Feb   | 120   |
| Mar   | 130   |
| ...   | ...   |
| Dec   | 150   |

We can decompose this data into trend, seasonality, and irregular components.

### 2. Moving Averages

Moving averages smooth out short-term fluctuations and highlight longer-term trends or cycles.

#### Simple Moving Average (SMA)

The SMA is calculated by averaging data points within a specified window.

*Example:*

For a 3-month moving average:

$$ \text{SMA}_t = \frac{Y_{t-2} + Y_{t-1} + Y_t}{3} $$

If sales for Jan, Feb, and Mar are 100, 120, and 130 respectively:

$$ \text{SMA}_{\text{Mar}} = \frac{100 + 120 + 130}{3} = 116.67 $$

#### Exponential Moving Average (EMA)

The EMA gives more weight to recent observations.

*Example:*

For a span $\alpha$ of 0.3:

$$ \text{EMA}_t = \alpha Y_t + (1 - \alpha) \text{EMA}_{t-1} $$

If the previous EMA was 110 and the current sales are 130:

$$ \text{EMA}_{\text{current}} = 0.3 \times 130 + 0.7 \times 110 = 116 $$

### 3. Autoregressive (AR) Models

Autoregressive models predict future values based on past values.

*Example:*

An AR(1) model uses one lag:

$$ Y_t = \phi Y_{t-1} + \epsilon_t $$

Where:
- $Y_t$ = current value
- $Y_{t-1}$ = previous value
- $\phi$ = autoregressive parameter
- $\epsilon_t$ = error term

If $\phi = 0.8$ and the previous value was 100:

$$ Y_t = 0.8 \times 100 + \epsilon_t $$

### 4. Moving Average (MA) Models

Moving Average models use past forecast errors to predict future values.

*Example:*

An MA(1) model uses one lagged error term:

$$ Y_t = \mu + \epsilon_t + \theta \epsilon_{t-1} $$

Where:
- $Y_t$ = current value
- $\mu$ = mean of the series
- $\epsilon_t$ = current error term
- $\epsilon_{t-1}$ = previous error term
- $\theta$ = MA parameter

If $\mu = 0$, $\theta = 0.5$, and the previous error was 20:

$$ Y_t = 0 + \epsilon_t + 0.5 \times 20 = \epsilon_t + 10 $$

### 5. ARMA and ARIMA Models

#### ARMA Models

ARMA models combine AR and MA models.

*Example:*

An ARMA(1,1) model:

$$ Y_t = \phi Y_{t-1} + \epsilon_t + \theta \epsilon_{t-1} $$

#### ARIMA Models

ARIMA models (Autoregressive Integrated Moving Average) include differencing to make the data stationary.

*Example:*

An ARIMA(1,1,1) model:

1. Differencing:

$$ \Delta Y_t = Y_t - Y_{t-1} $$

2. ARMA on differenced data:

$$ \Delta Y_t = \phi \Delta Y_{t-1} + \epsilon_t + \theta \epsilon_{t-1} $$

### 6. Seasonal Decomposition

Seasonal decomposition separates a time series into trend, seasonal, and residual components.

*Example:*

Using sales data:

1. **Trend**: Smooth the data to identify the trend.
2. **Seasonal**: Average the data across seasons.
3. **Residual**: Subtract trend and seasonal components from the original data.

### 7. Stationarity and Differencing

Stationarity implies that the statistical properties of the series do not change over time. Differencing can make a non-stationary series stationary.

*Example:*

Suppose the sales data show a trend. Differencing can remove this trend:

$$ \Delta Y_t = Y_t - Y_{t-1} $$

### 8. Autocorrelation and Partial Autocorrelation

#### Autocorrelation Function (ACF)

ACF measures the correlation between the time series and lagged versions of itself.

*Example:*

If sales data show correlation at lag 1:

$$ \text{ACF}(1) = \frac{\sum (Y_t - \bar{Y})(Y_{t-1} - \bar{Y})}{\sum (Y_t - \bar{Y})^2} $$

#### Partial Autocorrelation Function (PACF)

PACF measures the correlation between the time series and lagged versions of itself, controlling for shorter lags.

*Example:*

If sales data show partial correlation at lag 1:

$$ \text{PACF}(1) = \text{Correlation}(Y_t, Y_{t-1} | Y_{t-2}) $$

### 9. Forecasting

Time series models can be used to make forecasts.

*Example:*

Using an ARIMA model to forecast future sales:

1. Fit the model to historical data.
2. Use the model to generate future values.

### 10. Summary

Time series analysis provides tools for analyzing data collected over time. Understanding components, moving averages, AR, MA, ARMA, and ARIMA models, as well as stationarity, autocorrelation, and forecasting, is essential for effective time series analysis.

### Numerical Example

Let's apply a numerical example to an ARIMA(1,1,1) model.

Given monthly sales data for six months:

| Month | Sales |
|-------|-------|
| Jan   | 100   |
| Feb   | 120   |
| Mar   | 130   |
| Apr   | 115   |
| May   | 140   |
| Jun   | 150   |

1. **Differencing**:

$$ \Delta Y_{\text{Feb}} = 120 - 100 = 20 $$
$$ \Delta Y_{\text{Mar}} = 130 - 120 = 10 $$
$$ \Delta Y_{\text{Apr}} = 115 - 130 = -15 $$
$$ \Delta Y_{\text{May}} = 140 - 115 = 25 $$
$$ \Delta Y_{\text{Jun}} = 150 - 140 = 10 $$

2. **ARIMA Model**:

Fit an ARIMA(1,1,1) model to the differenced data.

$$ \Delta Y_t = \phi \Delta Y_{t-1} + \epsilon_t + \theta \epsilon_{t-1} $$

Assume $\phi = 0.5$ and $\theta = 0.3$ for simplicity.

For May (forecasting June):

$$ \Delta Y_{\text{May}} = 0.5 \times 25 + \epsilon_{\text{May}} + 0.3 \times \epsilon_{\text{Apr}} $$

Assume $\epsilon_{\text{Apr}} = -10$ and $\epsilon_{\text{May}} = 5$:

$$ \Delta Y_{\text{May}} = 12.5 + 5 + (-3) = 14.5 $$

Forecast for June:

$$ Y_{\text{Jun}} = Y_{\text{May}} + \Delta Y_{\text{May}} = 140 + 14.5 = 154.5 $$

Thus, the forecasted sales for June are approximately 154.5 units.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Load the dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Display first few rows
print(data.head())

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Passengers'], marker='o', linestyle='-')
plt.title('Monthly Airline Passengers')
plt.xlabel('Month')
plt.ylabel('Passengers (1000s)')
plt.grid(True)
plt.show()

# Decompose the time series into trend, seasonal, and residual components
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

plt.figure(figsize=(12, 8))
plt.subplot(411)
plt.plot(data.index, decomposition.observed, label='Observed', marker='o')
plt.legend(loc='upper left')
plt.subplot(412)
plt.plot(data.index, decomposition.trend, label='Trend', marker='o')
plt.legend(loc='upper left')
plt.subplot(413)
plt.plot(data.index, decomposition.seasonal, label='Seasonal', marker='o')
plt.legend(loc='upper left')
plt.subplot(414)
plt.plot(data.index, decomposition.resid, label='Residual', marker='o')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# ACF and PACF plots
plot_acf(data['Passengers'], lags=50)
plt.title('Autocorrelation Function (ACF)')
plt.xlabel('Lags')
plt.ylabel('ACF')
plt.show()

plot_pacf(data['Passengers'], lags=50)
plt.title('Partial Autocorrelation Function (PACF)')
plt.xlabel('Lags')
plt.ylabel('PACF')
plt.show()

# Fitting an ARIMA model
model = ARIMA(data['Passengers'], order=(2, 1, 2))  # Example order (p, d, q)
results = model.fit()

# Summary of the model
print(results.summary())


```
