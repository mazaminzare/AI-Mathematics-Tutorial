## Inferential Statistics

Inferential statistics involve making inferences about populations based on samples taken from them. This tutorial covers key concepts, mathematical background, and numerical examples.

### 1. Estimation

Estimation is the process of inferring the value of a population parameter based on a sample statistic. There are two main types of estimation:

- **Point Estimation**
- **Interval Estimation**

#### Point Estimation

Point estimation provides a single value as an estimate of a population parameter.

*Example:*

Given a sample mean ($\bar{x}$) as an estimate for the population mean ($\mu$):

Suppose we have a sample of test scores: 75, 80, 85, 90. The sample mean ($\bar{x}$) is calculated as:

$$ \bar{x} = \frac{75 + 80 + 85 + 90}{4} = 82.5 $$

So, the point estimate for the population mean ($\mu$) is 82.5.

**Key Properties:**

1. **Unbiasedness**: An estimator is unbiased if its expected value equals the true parameter value.
2. **Consistency**: An estimator is consistent if it converges to the true parameter value as the sample size increases.
3. **Efficiency**: Among unbiased estimators, the one with the smallest variance is the most efficient.

#### Interval Estimation

Interval estimation provides a range of values, called a confidence interval, within which the parameter is expected to lie.

*Example:*

Confidence Interval for the Mean (with known variance):

Suppose the population standard deviation ($\sigma$) is 10 and we have a sample of size 25 with a sample mean ($\bar{x}$) of 82.5. If we want to construct a 95% confidence interval for the population mean ($\mu$):

$$ \bar{x} \pm Z_{\alpha/2} \left( \frac{\sigma}{\sqrt{n}} \right) $$

For a 95% confidence level, $Z_{\alpha/2} = 1.96$ (from the standard normal distribution).

$$ 82.5 \pm 1.96 \left( \frac{10}{\sqrt{25}} \right) $$

$$ 82.5 \pm 1.96 \times 2 $$

$$ 82.5 \pm 3.92 $$

So, the 95% confidence interval for the population mean ($\mu$) is approximately [78.58, 86.42].

**Key Properties:**

1. **Confidence Level**: The probability that the interval contains the true parameter value (e.g., 95% confidence level).
2. **Margin of Error**: The maximum expected difference between the point estimate and the true parameter.
3. **Width of Interval**: Depends on the sample size, variability, and confidence level.

### 2. Hypothesis Testing

Hypothesis testing is a method of making decisions about population parameters based on sample data. It involves:

- **Null Hypothesis ($H_0$)**
- **Alternative Hypothesis ($H_a$)**
- **Test Statistic**
- **P-value**
- **Significance Level ($\alpha$)**
- **Decision Rule**

#### Steps in Hypothesis Testing

1. **State the Hypotheses**:

    - Null Hypothesis ($H_0$): Assumes no effect or no difference.
    - Alternative Hypothesis ($H_a$): Assumes an effect or a difference.

2. **Choose the Significance Level ($\alpha$)**: Common values are 0.05, 0.01.

3. **Select the Test Statistic**: Depending on the data and hypotheses, common tests include t-test, z-test, chi-square test.

4. **Compute the Test Statistic and P-value**:

    - *Example (t-test)*:

    Suppose we want to test if the average test score is significantly different from 80. We have a sample mean ($\bar{x}$) of 82.5, sample standard deviation ($s$) of 10, and a sample size of 25.

    $$ t = \frac{\bar{x} - \mu_0}{s / \sqrt{n}} = \frac{82.5 - 80}{10 / \sqrt{25}} $$

    $$ t = \frac{2.5}{2} = 1.25 $$

    - Compare the test statistic to critical values from the t-distribution table or use the p-value approach.

5. **Make a Decision**:

    - If p-value $\leq \alpha$, reject $H_0$.
    - If p-value $> \alpha$, fail to reject $H_0$.

**Key Properties:**

1. **Type I Error**: Rejecting $H_0$ when it is true (false positive).
2. **Type II Error**: Failing to reject $H_0$ when it is false (false negative).
3. **Power of the Test**: The probability of correctly rejecting a false $H_0$ (1 - Type II Error rate).

### 3. Regression Analysis

Regression analysis examines the relationship between variables. The most common type is linear regression.

#### Simple Linear Regression

Simple linear regression models the relationship between two variables by fitting a linear equation to observed data.

*Example:*

Suppose we want to predict students' test scores ($y$) based on the number of study hours ($x$). We have the following data:

| Study Hours ($x$) | Test Scores ($y$) |
|-------------------|-------------------|
| 2                 | 65                |
| 3                 | 70                |
| 4                 | 75                |
| 5                 | 80                |

The simple linear regression equation is:

$$ y = \beta_0 + \beta_1 x + \epsilon $$

Where:
- $y$ = Test Scores (dependent variable)
- $x$ = Study Hours (independent variable)
- $\beta_0$ = Intercept
- $\beta_1$ = Slope
- $\epsilon$ = Error term

**Key Properties:**

1. **Least Squares Estimation**: Minimizes the sum of the squared differences between observed and predicted values.
2. **Coefficient of Determination ($R^2$)**: Indicates the proportion of variance in the dependent variable explained by the independent variable.
3. **Assumptions**: Linearity, independence, homoscedasticity, normality of errors.

#### Multiple Linear Regression

Multiple linear regression models the relationship between one dependent variable and two or more independent variables.

*Example:*

Suppose we want to predict students' test scores ($y$) based on both study hours ($x_1$) and attendance ($x_2$). We have the following data:

| Study Hours ($x_1$) | Attendance ($x_2$) | Test Scores ($y$) |
|---------------------|---------------------|-------------------|
| 2                   | 90                  | 65                |
| 3                   | 95                  | 70                |
| 4                   | 100                 | 75                |
| 5                   | 105                 | 80                |

The multiple linear regression equation is:

$$ y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \epsilon $$

Where:
- $y$ = Test Scores (dependent variable)
- $x_1$ = Study Hours (independent variable)
- $x_2$ = Attendance (independent variable)
- $\beta_0$ = Intercept
- $\beta_1$, $\beta_2$ = Slopes for $x_1$ and $x_2$
- $\epsilon$ = Error term

**Key Properties:**

1. **Interpretation of Coefficients**: Each $\beta_i$ represents the change in $y$ for a one-unit change in $x_i$, holding other variables constant.
2. **Multicollinearity**: When independent variables are highly correlated, it can affect the stability and interpretation of the coefficients.
3. **Model Fit**: Evaluated using $R^2$, adjusted $R^2$, and other goodness-of-fit measures.

### 4. Summary

Inferential statistics allow us to make conclusions about population parameters based on sample data. Estimation and hypothesis testing provide frameworks for inference, while regression analysis helps in understanding relationships between variables. Mastery of these concepts enables effective data analysis and decision-making.

### 4. Analysis of Variance (ANOVA)

ANOVA is used to compare the means of two or more groups to determine if there is a statistically significant difference between them.

*Example:*

Suppose we have three groups of students with test scores:

Group 1: 75, 80, 85

Group 2: 70, 75, 80

Group 3: 65, 70, 75

We want to test if there is a significant difference in the mean test scores between these groups.

**Key Properties:**

1. **Between-group Variation**: Variability in the means of different groups.
2. **Within-group Variation**: Variability within each group.
3. **F-statistic**: Ratio of between-group variation to within-group variation.

### 5. Chi-Square Test

The chi-square test is used to determine whether there is a significant association between categorical variables.

*Example:*

Suppose we have survey data on students' preferences:

|           | Math | Science | Arts |
|-----------|------|---------|------|
| Male      | 30   | 20      | 10   |
| Female    | 25   | 15      | 20   |

We want to test if there is a relationship between gender and subject preference.

**Key Properties:**

1. **Expected Frequencies**: Frequencies expected under the null hypothesis of independence.
2. **Chi-Square Statistic**: Measure of the discrepancy between observed and expected frequencies.
3. **Degrees of Freedom**: Determined by the number of categories in the variables.

### 6. Non-Parametric Tests

Non-parametric tests are used when data do not meet the assumptions required for parametric tests (e.g., normal distribution).

*Example:*

Wilcoxon Signed-Rank Test:

Suppose we have data on the scores of students before and after a tutoring program:

| Student | Before | After |
|---------|--------|-------|
| A       | 60     | 70    |
| B       | 65     | 75    |
| C       | 70     | 68    |
| D       | 75     | 72    |
| E       | 80     | 78    |

We want to test if there is a significant difference in scores before and after tutoring.

**Key Properties:**

1. **Ranking of Data**: Observations are ranked rather than using the actual values.
2. **Test Statistic**: Calculated based on ranks or differences in ranks.
3. **Assumptions**: Less stringent assumptions compared to parametric tests.

### 7. Confidence Intervals for Proportions

Confidence intervals can also be constructed for population proportions.

*Example:*

Suppose in a survey, 150 out of 200 people prefer a new product. The sample proportion ($\hat{p}$) is:

$$ \hat{p} = \frac{150}{200} = 0.75 $$

To construct a 95% confidence interval for the population proportion ($p$):

$$ \hat{p} \pm Z_{\alpha/2} \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} $$

For a 95% confidence level, $Z_{\alpha/2} = 1.96$.

$$ 0.75 \pm 1.96 \sqrt{\frac{0.75(1 - 0.75)}{200}} $$

$$ 0.75 \pm 1.96 \sqrt{\frac{0.1875}{200}} $$

$$ 0.75 \pm 1.96 \sqrt{0.0009375} $$

$$ 0.75 \pm 1.96 \times 0.0306 $$

$$ 0.75 \pm 0.06 $$

So, the 95% confidence interval for the population proportion ($p$) is approximately [0.69, 0.81].

### 8. Hypothesis Testing for Proportions

Hypothesis testing can also be applied to population proportions.

*Example:*

Suppose we want to test if the proportion of people who prefer a new product is 0.70. We have a sample proportion ($\hat{p}$) of 0.75 from a sample size of 200.

1. **State the Hypotheses**:

    - $H_0$: $p = 0.70$
    - $H_a$: $p \neq 0.70$

2. **Choose the Significance Level ($\alpha$)**: 0.05.

3. **Select the Test Statistic**:

$$ Z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0 (1 - p_0)}{n}}} $$

4. **Compute the Test Statistic and P-value**:

$$ Z = \frac{0.75 - 0.70}{\sqrt{\frac{0.70 (1 - 0.70)}{200}}} $$

$$ Z = \frac{0.05}{\sqrt{\frac{0.21}{200}}} $$

$$ Z = \frac{0.05}{\sqrt{0.00105}} $$

$$ Z = \frac{0.05}{0.0324} $$

$$ Z \approx 1.54 $$

Using standard normal distribution tables or software, we find the p-value.

5. **Make a Decision**:

If p-value $\leq \alpha$, reject $H_0$.
If p-value $> \alpha$, fail to reject $H_0$.

### 9. Correlation

Correlation measures the strength and direction of the linear relationship between two variables.

*Example:*

Suppose we have data on the number of study hours and test scores:

| Study Hours ($x$) | Test Scores ($y$) |
|-------------------|-------------------|
| 2                 | 65                |
| 3                 | 70                |
| 4                 | 75                |
| 5                 | 80                |

The correlation coefficient ($r$) is calculated as:

$$ r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum (x_i - \bar{x})^2 \sum (y_i - \bar{y})^2}} $$

**Key Properties:**

1. **Range**: $-1 \leq r \leq 1$.
2. **Direction**: Positive $r$ indicates a positive relationship, negative $r$ indicates a negative relationship.
3. **Strength**: Values close to -1 or 1 indicate a strong relationship, values close to 0 indicate a weak relationship.

```python
import numpy as np
# Sample data
sample_data = np.array([75, 80, 85, 90])

# Point estimate for the mean
point_estimate = np.mean(sample_data)
print(f'Point Estimate (Mean): {point_estimate}')
```

```python
# Interval Estimation

import scipy.stats as stats

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)
pop_std = 10  # Known population standard deviation
confidence_level = 0.95

# Z value for 95% confidence
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of error
margin_of_error = z_value * (pop_std / np.sqrt(sample_size))

# Confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print(f'95% Confidence Interval: {confidence_interval}')
```

```python
# Hypothesis Testing
sample_mean = np.mean(sample_data)
pop_mean = 80
sample_std = np.std(sample_data, ddof=1)
sample_size = len(sample_data)

# t-test for one sample
t_statistic, p_value = stats.ttest_1samp(sample_data, pop_mean)
print(f'T-statistic: {t_statistic}, P-value: {p_value}')
```

```python
# Regression Analysis
import statsmodels.api as sm

# Sample data
study_hours = np.array([2, 3, 4, 5])
test_scores = np.array([65, 70, 75, 80])

# Add a constant to the independent variable
X = sm.add_constant(study_hours)
model = sm.OLS(test_scores, X).fit()
print(model.summary())
```

```python
# Multiple Linear Regression
# Sample data
study_hours = np.array([2, 3, 4, 5])
attendance = np.array([90, 95, 100, 105])
test_scores = np.array([65, 70, 75, 80])

# Combine independent variables into a single matrix
X = np.column_stack((study_hours, attendance))
X = sm.add_constant(X)
model = sm.OLS(test_scores, X).fit()
print(model.summary())
```

```python
# Analysis of Variance (ANOVA)
# Sample data
group1 = np.array([75, 80, 85])
group2 = np.array([70, 75, 80])
group3 = np.array([65, 70, 75])

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)
print(f'F-statistic: {f_statistic}, P-value: {p_value}')
```

```python
# Chi-Square Test
# Sample data
observed = np.array([[30, 20, 10],
                     [25, 15, 20]])

# Perform chi-square test
chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed)
print(f'Chi2-statistic: {chi2_statistic}, P-value: {p_value}')
```

```python
# Non-Parametric Tests
# Wilcoxon Signed-Rank Test
# Sample data
before = np.array([60, 65, 70, 75, 80])
after = np.array([70, 75, 68, 72, 78])

# Perform Wilcoxon signed-rank test
statistic, p_value = stats.wilcoxon(before, after)
print(f'Statistic: {statistic}, P-value: {p_value}')
```

```python
# Confidence Intervals for Proportions
# Sample data
successes = 150
sample_size = 200
sample_proportion = successes / sample_size
confidence_level = 0.95

# Z value for 95% confidence
z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

# Margin of error
margin_of_error = z_value * np.sqrt((sample_proportion * (1 - sample_proportion)) / sample_size)

# Confidence interval
confidence_interval = (sample_proportion - margin_of_error, sample_proportion + margin_of_error)
print(f'95% Confidence Interval for Proportion: {confidence_interval}')
```

```python
#Hypothesis Testing for Proportions
# Sample data
successes = 150
sample_size = 200
sample_proportion = successes / sample_size
pop_proportion = 0.70

# Test statistic
z_statistic = (sample_proportion - pop_proportion) / np.sqrt((pop_proportion * (1 - pop_proportion)) / sample_size)
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

print(f'Z-statistic: {z_statistic}, P-value: {p_value}')
```

```python
# Correlation
# Sample data
study_hours = np.array([2, 3, 4, 5])
test_scores = np.array([65, 70, 75, 80])

# Compute correlation coefficient
correlation_coefficient, p_value = stats.pearsonr(study_hours, test_scores)
print(f'Correlation Coefficient: {correlation_coefficient}, P-value: {p_value}')
```
